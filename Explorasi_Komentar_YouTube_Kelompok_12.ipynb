```python
import re
import string

def clean_text(text):
    # menghapus emoticon
    text = text.encode('ascii', 'ignore').decode('ascii')
    # menghilangkan Nama Akun
    text = re.sub('@[^\s]+', '', text)
    # menggantu baris baru dengan spasi
    text = re.sub("\n", " ", text)
    # menghilangkan punctuation
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    # menghilangkan angka
    text = re.sub('\w*\d\w*', '', text)
    # menghilangkan URL
    text = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
    text = re.sub(r'(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w\.-]*)*\/?\s', '', text)
    # menghilangkan Hastag
    text = re.sub('#[^\s]+', '', text)
    # menghilangkan Huruf Tunggal
    text = re.sub(r'\b[a-zA-Z]\b','', text)

    return text
```

***Cleansing Data***

```python
# Membersihkan teks pada kolom 'textDisplay' dan menyimpannya pada kolom baru 'textDisplay_clean'
df['textDisplay_clean'] = df['textDisplay'].apply(clean_text)

# Menampilkan hasil
df[['textDisplay', 'textDisplay_clean']].head()
```

***Case Folding***

```python
# Melakukan case folding pada kolom 'review'
df['textDisplay_casefolding'] = df['textDisplay_clean'].str.lower()

# Menampilkan hasil
df[['textDisplay_clean', 'textDisplay_casefolding']].head()
```

***Stopwords Removal***

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = stopwords.words('indonesian')
df['textDisplay_stopword'] = df['textDisplay_casefolding'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df[['textDisplay_casefolding', 'textDisplay_stopword']].head(20)
```

***Hitung dan Hapus baris data yang kosong***

```python
empty_strings_count = df['textDisplay_stopword'].apply(lambda x: len(x) == 0).sum()
print(f"Jumlah string kosong: {empty_strings_count}")
```

```python
# Hapus baris dengan string atau list kosong di kolom 'textDisplay_stopword'
df = df[df['textDisplay_stopword'].apply(lambda x: len(x) > 0)]

# Periksa kembali jumlah baris setelah penghapusan
print(f"Jumlah baris setelah penghapusan: {len(df)}")
```

***Tokenizing***

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

df['textDisplay_token'] = df['textDisplay_stopword'].apply(lambda x: word_tokenize(x))
df[['textDisplay_stopword', 'textDisplay_token']].head(10)
```

***Stemming***

```python
# Inisialisasi stemmer untuk bahasa indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Melakukan stemming pada kolom 'toxic'
df['textDisplay_stemming'] = df['textDisplay_token'].apply(lambda x: [stemmer.stem(word) for word in x])

# Menampilkan hasil
df[['textDisplay_token', 'textDisplay_stemming']].head(30)
```
